```python
li_candidate_h24 = list()
for space in S:
    m = np.linalg.matrix_rank(space[0], tol=None)
    l = np.linalg.matrix_rank(space[1], tol=None)
    if m == 2 and l == 2:
        M = np.hstack((space[0], space[1]))
        print(Matrix(M).rref())
        print("-----------")
        li_candidate_h24.append(M)
```
