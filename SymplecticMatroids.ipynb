# ～～ Symplectic $U_{k, n}$ 表現探索の旅 ～～

## $U_{k,n}$ 表現の全探索

```python
import numpy as np
import itertools

import SymplecticMatroids


p = 5
n = 4
k = 2
```

```python
def det(M):
    M = [row[:] for row in M] # make a copy to keep original M unmodified
    N, sign, prev = len(M), 1, 1
    for i in range(N-1):
        if M[i][i] == 0: # swap with another row having nonzero i's elem
            swapto = next( (j for j in range(i+1,N) if M[j][i] != 0), None )
            if swapto is None:
                return 0 # all M[*][i] are zero => zero determinant
            M[i], M[swapto], sign = M[swapto], M[i], -sign
        for j in range(i+1,N):
            for k in range(i+1,N):
                assert ( M[j][k] * M[i][i] - M[j][i] * M[i][k] ) % prev == 0
                M[j][k] = ( M[j][k] * M[i][i] - M[j][i] * M[i][k] ) // prev
        prev = M[i][i]
    return sign * M[-1][-1]
```

```python
Fk = list(itertools.product(range(p), repeat=k))
Fk = np.array(list(map(lambda x:np.array(x), Fk)))

d_det = dict()
for t_idx in itertools.combinations_with_replacement(range(len(Fk)), k):
    d_det[t_idx] = det(Fk[list(t_idx)])%p
```

```python
matrices = itertools.combinations_with_replacement(range(len(Fk)),n)
matrices = list(map(lambda x:np.array(x) , matrices))
matrices = matrices[1:] # 0行列はいらない

subset_indeces = tuple(itertools.combinations(range(n),k))

v_U_kn = list()
for matrix in matrices:
    flag = True
    for idx in subset_indeces:
        if d_det[tuple(matrix[list(idx)])] == 0:
            flag = False
            break
    if flag:
        v_U_kn.append(matrix)
```

```python
print(len(v_U_kn))
print(Fk[list(v_U_kn[5])].T)
```

```python
U_kn = [np.array([np.array([0 for _ in range(k)]) for i in range(n)]) for j in range(len(v_U_kn))]

for u in range(len(U_kn)):
    for idx in range(n):
        U_kn[u][idx] = Fk[v_U_kn[u][idx]]
```

```python
print(len(U_kn))
print(U_kn[5].T)
```

```python
candidate0 = list()
for idxA, idxB in itertools.combinations(range(len(U_kn)), 2):
    ABt = (U_kn[idxA].T@U_kn[idxB])%p
    if np.array_equal(ABt, ABt.T):
        candidate0.append((v_U_kn[idxA],v_U_kn[idxB]))
```

```python
print("A")
print(Fk[list(candidate0[5][0])].T)

print("B")
print(Fk[list(candidate0[5][1])].T)

print("ABt")
print(Fk[list(candidate0[5][0])].T@Fk[list(candidate0[5][1])]%p)

print(len(candidate0))
```

```python
admissible_subsets = SymplecticMatroids.make_admissible_subsets(n,k)

for X in admissible_subsets:
    for e in range(len(X)):
        if X[e] > 0:
            X[e] = X[e] - 1
        else:
            X[e] = n-1 - X[e]
    X.sort()
```

```python
candidate1 = list()
for AB in candidate0:
    flag = True
    A_B = np.hstack(AB)
    for X in admissible_subsets:
        if d_det[tuple(sorted(A_B[X]))] == 0:
            flag = False
            break
    if flag:
        candidate1.append(AB)
```

```python
admissible_subsets = set(map(tuple, admissible_subsets))

subsets = set(itertools.combinations(range(2*n), k))
non_admissible_subsets = subsets - admissible_subsets

candidate2 = list()
for AB in candidate1:
    A_B = np.hstack(AB)
    flag = True
    for X in non_admissible_subsets:
        if  d_det[tuple(sorted(A_B[list(X)]))] != 0:
            flag = False
            break

    if flag:
        candidate2.append((A,B))
```

```python
candidate3 = list()
for A, B in candidate2:
    At = Fk[A]
    Bt = Fk[B]

    flag0 = False # [ある列i, i* が存在して独立]
    for i in range(n):
        flag1 = True # [任意のスカラー倍でベクトルが等しくない(独立)]
        for k in range(p):
            if np.array_equal(At[i],k*Bt[i]%p):
               flag1 = False
               break

        if flag1:
            flag0 = True
            break
    
    if flag0:
        candidate3.append((A,B))
```
