# ～～ Symplectic $U_{k, n}$ 表現探索の旅 ～～

## $U_{k,n}$ 表現の全探索

```python
import numpy as np
import itertools

p = 5
n = 4
k = 2
```

```python
def det(M):
    M = [row[:] for row in M] # make a copy to keep original M unmodified
    N, sign, prev = len(M), 1, 1
    for i in range(N-1):
        if M[i][i] == 0: # swap with another row having nonzero i's elem
            swapto = next( (j for j in range(i+1,N) if M[j][i] != 0), None )
            if swapto is None:
                return 0 # all M[*][i] are zero => zero determinant
            M[i], M[swapto], sign = M[swapto], M[i], -sign
        for j in range(i+1,N):
            for k in range(i+1,N):
                assert ( M[j][k] * M[i][i] - M[j][i] * M[i][k] ) % prev == 0
                M[j][k] = ( M[j][k] * M[i][i] - M[j][i] * M[i][k] ) // prev
        prev = M[i][i]
    return sign * M[-1][-1]
```

```python
Fk = list(itertools.product(range(p), repeat=k))

matrices = itertools.combinations_with_replacement(Fk,n)
matrices = list(map(lambda x:np.array(list(x)), matrices))
matrices = matrices[1:] # 0行列はいらない

subset_indeces = itertools.combinations(range(n),k)
subset_indeces = all_col = list(map(list, subset_indeces))

U = list()
for matrix in matrices:
    flag = True
    for idx in subset_indeces:
        if det(matrix[idx])%p == 0:
            flag = False
            break
    if flag:
        U.append(matrix.T)
```

```python
admissible_subsets = SymplecticMatroids.make_admissible_subsets(n,k)

for X in admissible_subsets:
    for e in range(len(X)):
        if X[e] > 0:
            X[e] = X[e] - 1
        else:
            X[e] = n-1 - X[e]

admissible_subsets = set(map(lambda x:tuple(sorted(x)), admissible_subsets))
admissible_subsets
```

```python
candidate = list()
count = 0
limit = 1000000000
for idxA in range(len(U)-1):
    for idxB in range(idxA+1,len(U)):
        ABt = U[idxA]@(U[idxB].T)
        if np.array_equal(ABt, ABt.T):
            candidate.append((U[idxA],U[idxB]))
            count += 1
    
        if count == limit:
            break
    if count == limit:
        break
```
